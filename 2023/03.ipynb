```python
sample = """467..114..
...*......
..35..633.
......#...
617*......
.....+.58.
..592.....
......755.
...$.*....
.664.598.."""

import re 

def solve(input_string):
    lines = input_string.split('\n')
    xmax = len(lines[0])
    ymax = len(lines)
    valid_symbols = '$*?+!§%&#/()*+-,\'"'

    def is_symbol(char: str):
        if char=='.': return False
        if char.isalpha() or char.isdigit(): return False
        return True
    
    def get_value(x,y):
        return lines[y][x]

    def get_adjacent_indices(x,y):
        candidates = [
            [x-1,y-1],
            [x,y-1],
            [x+1,y-1],
            [x-1,y],
            [x+1,y],
            [x-1,y+1],
            [x,y+1],
            [x+1,y+1]
        ]
        return list(filter(lambda k: 0 <=k[0] <xmax and 0<=k[1]<ymax,candidates))

    
    numbers = []
    for y, line in enumerate(lines):
        for m in re.finditer(r'\d+',line):
            keep_searching=True
            for x in range(m.start(), m.end()):
                checks = get_adjacent_indices(x, y)
                for c in checks:
                    val = get_value(*c)
                    # print(m[0],*c,val)
                    if is_symbol(val):
                        keep_searching=False
                        numbers.append(int(m[0]))
                        break
                if not keep_searching:
                    break
    print(numbers)
    return sum(numbers)

solve(sample)
```

```python
sample2="""467..114..
...*......
..35..633.
......#...
617*......
.....+.58.
..592.....
......755.
...$.*....
.664.598.."""



import numpy as np
import itertools
from re import Match
import functools

def solve2(input_string):
    lines = input_string.split('\n')
    xmax = len(lines[0])
    ymax = len(lines)

    def is_symbol(char: str):
        if char=='.': return False
        if char.isalpha() or char.isdigit(): return False
        return True
    
    def get_value(x,y):
        return lines[y][x]

    def get_adjacent_indices(x,y):
        candidates = [
            [x-1,y-1],
            [x,y-1],
            [x+1,y-1],
            [x-1,y],
            [x+1,y],
            [x-1,y+1],
            [x,y+1],
            [x+1,y+1]
        ]
        return list(filter(lambda k: 0 <=k[0] <xmax and 0<=k[1]<ymax,candidates))

    
    def match_is_next_to(m:Match,y1,x2,y2):
        #print(m.span())
        #print(range(*m.span()))
        #print(y1)
        return any((is_adjacent_to(x1,y1,x2,y2) for x1 in range(*m.span()) ))

    def is_adjacent_to(x1,y1,x2,y2):
        dx=abs(x1-x2)
        dy=abs(y1-y2)
        return dx<=1 and dy <=1 and (dx+dy) != 0 

    def get_adjacent_numbers(x,y):
        y_vals = set(np.array(get_adjacent_indices(x,y))[:,1])
        #print(y_vals)
        iterator_list = []
        for yi in y_vals:
            candidates = re.finditer(r'\d+',lines[yi])
            iterator_list.append(filter(lambda m: match_is_next_to(m,yi,x,y), candidates))
        matching = list(itertools.chain(*iterator_list))
        #print(matching)
        if len(matching) == 2:
            return functools.reduce(lambda a,b: a*b,map(lambda x: int(x[0]), matching))
        return 0

    assert is_adjacent_to(1,1,1,0)
    assert not is_adjacent_to(1,1,5,2)
    assert not is_adjacent_to(1,1,1,1)
    assert is_adjacent_to(1,1,0,0)
 
    numbers = []
    for y, line in enumerate(lines):
        for m in re.finditer(r'\*',line):
            x = m.start()
            numbers.append(get_adjacent_numbers(x,y))
    return sum(numbers)

solve2(sample2)
```
